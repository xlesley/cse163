```python

# Define constants
IMAGE_SIZE = 128
NUM_CLASSES = 257
BATCH_SIZE = 32

# Define function to preprocess images
def preprocess_image(image_path):
    # Load image
    image = Image.open(image_path)
    # Resize image
    image = image.resize((IMAGE_SIZE, IMAGE_SIZE))
    # Convert image to NumPy array
    image_array = np.array(image)
    # Normalize pixel values to range [0, 1]
    image_array = image_array / 255.0
    return image_array

# Define function to create TensorFlow dataset
def create_dataset(image_dir):
    # Create list of image paths
    image_paths = [os.path.join(image_dir, file) for file in os.listdir(image_dir)]
    # Create list of corresponding labels
    labels = [int(os.path.basename(path).split("_")[0]) for path in image_paths]
    # Preprocess images and convert labels to one-hot encoding
    images = np.array([preprocess_image(path) for path in image_paths])
    labels = tf.one_hot(labels, NUM_CLASSES)
    # Create TensorFlow dataset
    dataset = tf.data.Dataset.from_tensor_slices((images, labels))
    # Shuffle and batch dataset
    dataset = dataset.shuffle(len(images)).batch(BATCH_SIZE)
    return dataset

# Create training dataset
train_dataset = create_dataset("caltech256/train")

# Create validation dataset
val_dataset = create_dataset("caltech256/val")
```
